```python
import sympy
#from sympy import*          #### another way

x, y, z = symbols("X, y, z")
f = 2*x**3 + x**2 + 3*x + 5
res1 = diff(f, x)
res2 = diff(f, x, 2)
print(res1, "\n",res2)
print(f.subs(x, 4))
print(res2.subs(x, 4).evalf())

s = 3*x**2 + 4*x*y - z

res3 = diff(s, x, 2)
dif3 = 2*x+y+z
print(res3)
print(dif3.subs([(x, 1), (y,2), (z, 3)]))
```

```python
f = lambdify(x, f)
res1 = lambdify(x, res1)
res2 = lambdify(x, res2)
print(f(-1))
print(res1(3))
print(res2(3))
```
