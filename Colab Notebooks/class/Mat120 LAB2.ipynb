```python
import sympy as smp
import numpy as np
import scipy as sp
from scipy.integrate import quad

x, a= smp.symbols("x, a")    #### define x as symbol

f= smp.sin(x)**3*smp.exp(-5*x)         # exp means e in math
smp.integrate(f, x).simplify()
```

```python
f = smp.exp(-x)*x**2*smp.cos(x)
smp.integrate(f, x).simplify()
```

```python
f = (1+smp.sqrt(x))**smp.Rational(1, 3)/smp.sqrt(x)
smp.integrate(f, (x, 0, smp.oo)).simplify().evalf()
```

```python
f = smp.exp(-smp.sin(x))
smp.integrate(f, (x, 1, 2)).simplify()
```

```python
f = lambda x: np.exp(-np.sin(x))
quad(f, 1, 2)
```
