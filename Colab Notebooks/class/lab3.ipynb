```python
############################################
#### pretty print .....  :)
import sympy as sy
x = sy.symbols("x")
y = sy.Function("y")
y1 = sy.Derivative(y(x), x)

sy.init_printing()

eqdiff = y1 + (y(x)**2)/x**2
sol = sy.dsolve(eqdiff, y(x))
sy.pprint(sol)
```

```python
sol = sy.dsolve(eqdiff, y(x), ics = {y(1): '1/2'})
sy.pprint(sol)
```

```python
 from sympy.plotting import plot
 xtol = 1e-2
 p1 = plot(sol.rhs, (x, -1, 1/3-xtol), ylim = [-10, 10])
 p2 = plot(sol.rhs, (x, 1/3+xtol, 1), ylim = [-10, 10])
 p1.extend(p2)
 p1.show()
```
