```python
# 3 a
import sympy as smp
import numpy as np
import scipy as sp
from scipy.integrate import quad

x = smp.symbols("x")    #### define x as symbol
a,b = smp.symbols('a, b', real=True, positive=True)

f = smp.sin(a*x)*smp.exp(-b*x)         # exp means e in math
smp.integrate(f, x)
```
