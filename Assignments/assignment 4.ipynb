```python
####  Question 1
import sympy as sy
sy.init_printing()
from sympy.plotting import plot

x = sy.symbols('x')
y = sy.Function('y')
y1 = sy.Derivative(y(x), x)

y2 = sy.Derivative(y(x), x, x)


eqdiff = y2 + 16*y(x)


eq=sy.dsolve(eqdiff, y(x))

sy.pprint(eq)
```

```python
######  Question 2
eqdiff = y2 - 6*y1 + 5*y(x)

eq=sy.dsolve(eqdiff, y(x))

sy.pprint(eq)
```

```python
######  Question 3
eq_1=sy.dsolve(sy.Derivative(y(x), x, x) -sy.Derivative(y(x),x) + 6*y(x), y(x))
sy.pprint(eq_1)
k=sy.Derivative(y(x), x, x) -sy.Derivative(y(x),x) + 6*y(x)
p=y(x).diff(x)
eq_2=sy.dsolve(k, ics = {y(0):1,p.subs(x,0):1})
sy.pprint(eq_2)
```
