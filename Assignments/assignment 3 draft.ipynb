```python
#########################  Task 1  #######################
import sympy as sy
from sympy.plotting import plot
sy.init_printing() # initializing printing

t = sy.symbols('t')
y = sy.Function('y')

y1 = sy.Derivative(y(t), t)
eqdiff =y1 -  2*t*(25 - y(t))
sol = sy.dsolve(eqdiff, y(t))

sy.pprint(sol)
```

```python
############################  Task 2  ##############################
x = sy.symbols('x')
y = sy.Function('y')
y1 = sy.Derivative(y(x), x)
eqdiff = y1 - 2*x
sol = sy.dsolve(eqdiff, y(x), ics = {y(0): '2'})
sy.pprint(sol)
```

```python
############################  Question 2 #############################
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
def model(v, t):
    dvdt = 10-v**2
    return dvdt
t = np.linspace(1, 10, 1001)
y1 = 0
sol = odeint(model, y1, t)
plt.plot(t,sol, label = 'y')
plt.legend()
plt.savefig('num.svg')
plt.show()
```

```python
###########################  Question 3 ################################ 
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
l1 = [-4,-2,1,2,3]
for i in l1:
  def model(v, t):
      dvdt = 10-(i*(v**2))
      return dvdt
t = np.linspace(1, 10, 1001)
y1 = 0
sol = odeint(model, y1, t)
plt.plot(t,sol, label = 'y')
plt.legend()
plt.savefig('num.svg')
plt.show()
```
