**Question 1**

```python
from sympy import*
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy import ndimage


x, y, z = symbols("X, y, z")

def f1(x):
  f = 9*x**3-7*x**2+3*x+10
  return f
def f2(x):
  f = np.log(x)
  return f
def f3(x):
  f = np.sin(x)
  return f
def f4(x):
  f = np.cos(x)
  return f
def f5(x):
  f = np.tan(x)
  return f

####################################

df = diff(f1(x),x,1)
print("Differenciation:",df)

df = lambdify(x, df)

range = np.arange(-5, 6)
df_range = df(range)

print(df_range)
print("Maximum:",scipy.ndimage.maximum(df_range))
print("Minimum",scipy.ndimage.minimum(df_range))


temp = np.linspace(-5,6)
plt.plot(temp, f1(temp))
```
